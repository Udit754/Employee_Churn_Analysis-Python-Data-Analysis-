```python
# Countplot
plt.figure(figsize=(10, 5))
sns.countplot(x='numberOfProjects', hue='left', data=df)
plt.show()
```

#### Inference:

People with number of projects 2 and 7 are leaving most.

#### Suggestion:

* All the employees should be assigned 3-5 projects on an average.

* It should be re-distributed among employees working on 2 projects from 7 projects.

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='timeSpent.company', hue='left', data=df)
plt.show()
```

#### Inference:

Employees are leaving the organization after 3-5 years.

Those who completed more than 6 years in the company do not leave the company.

#### Suggestions:

Once they complete 3-5years, they should be given rewards(hike, better job position, incentives/increments).

# Conclusion

1) From Hr dept maximum people left then after that accounting , technical,support.

2) People with low salary are leaving so you can provide them some hike to stop them from leaving.

3) Project / Work is not equally distributed , distribute the project equally among all the employees.

4) Reasons for leaving company by employees who had spent fairly good time in between 3 years to 6 years may be either due to not getting promotion or getting salary hike as expected or combination of both.

Hence,the company should investigate further to know reasons behind for leaving by employees and approprate steps to be taken up such provide promotion and salary hike for achievers and upskilling for underperformers.
